```python
# Merge the first two columns of csv for all the files in a directory

# Get all files in directory
import os
dir = "/Users/akashanand/repo/marketmaestro/maestro/infra/data/backtrader/"
files = os.listdir(dir)
files[0]
```

```python
import pandas as pd
for name in files:
    fname = dir + name
    f = open(fname, 'r')
    print(fname)
    df = pd.read_csv(f)
    f.close()
    if 'datetime' in df.columns:
        continue
    df['datetime'] = df['date'] + " " + df['time']
    df = df.drop(['date', 'time'], axis=1)
    # Change order of columns
    cols = df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    df = df[cols]
    df.to_csv(fname, index=False)
```
